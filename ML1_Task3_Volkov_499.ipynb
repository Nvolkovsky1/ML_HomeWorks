```python
from sklearn import model_selection, datasets, metrics, neighbors 
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
import scipy.stats as st
from scipy import optimize 

%pylab inline
```

```python
#сгенерируем 500 точек
sample_size = 500

uniform = st.uniform(loc = 1, scale = 51)
x_exp = uniform.rvs(size=sample_size)
norm = st.norm(loc=0, scale=0.2)
noise = norm.rvs(size=sample_size)
y_exp = 0.5 * x_exp + np.ones(500) + noise
```

```python
#разделим выборку на тренировочную и тестовую части 
X_train, X_test, Y_train, Y_test = train_test_split(x_exp, y_exp, test_size=0.2, random_state=9)
```

```python
#оценим параметры методом MSE
def MSE(parameters):
    sug_sol = np.array([parameters[0]*arg + parameters[1] for arg in X_train])
    return float(sum((Y_train - sug_sol)**2)) / len(X_train)

res = optimize.minimize(MSE, [1, 1])
estimated_par = res.x
k_est = estimated_par[0]
b_est = estimated_par[1]
print k_est, b_est
```

```python
def get_linear_dots(k, b, begin, end):
    x = []
    step = float(end - begin) / 1000
    for iteration in xrange(1000):
        x.append(begin + step * iteration)
    y = [arg * k + b for arg in x]
    x = np.array(x)
    y = np.array(y)
    return [x, y]
    
```

```python
#Визуализируем выборку и прямую, полученную методом MSE 
plt.figure(figsize = (15,10))
plt.scatter(x_exp, y_exp, s = 1)
plt.xlim(0, 55)
plt.ylim(0, 30)

x_est = get_linear_dots(k_est, b_est, 0, 53)[0]
y_est = get_linear_dots(k_est, b_est, 0, 53)[1]
plt.plot(x_est, y_est, color = 'green')
```

```python
#добавим выбросы и посмотрим как сработает оценка методом MSE и MAE
new_x = []
for iterator in xrange(0,75):
    new_x.append(1 + float(50) / 75 * iterator)
new_y = (-1) * np.ones(75) + noise[:75]

X_train = np.concatenate((x_exp, new_x))
Y_train = np.concatenate((y_exp, new_y))
```

```python
def MSE(parameters):
    sug_sol = np.array([parameters[0]*arg + parameters[1] for arg in X_train])
    return float(sum((Y_train - sug_sol)**2)) / len(X_train)

res = optimize.minimize(MSE, [1, 1])
estimated_par = res.x
k_est_MSE = estimated_par[0]
b_est_MSE = estimated_par[1]

print k_est_MSE, b_est_MSE
```

```python
def MAE(parameters):
    sug_sol = np.array([parameters[0]*arg + parameters[1] for arg in X_train])
    return float(sum(abs(Y_train - sug_sol))) / len(X_train)

res = optimize.minimize(MAE, [1, 1])
estimated_par = res.x
k_est_MAE = estimated_par[0]
b_est_MAE = estimated_par[1]

print k_est_MAE, b_est_MAE
```

```python
plt.figure(figsize = (15,10))
plt.scatter(X_train, Y_train)
plt.xlim(0, 55)
plt.ylim(-5, 30)

x_est_MSE = get_linear_dots(k_est_MSE, b_est_MSE, 0, 54)[0]
y_est_MSE = get_linear_dots(k_est_MSE, b_est_MSE, 0, 54)[1]
plt.plot(x_est_MSE, y_est_MSE, color = 'green', linewidth = 3)

x_est_MAE = get_linear_dots(k_est_MAE, b_est_MAE, 0, 54)[0]
y_est_MAE = get_linear_dots(k_est_MAE, b_est_MAE, 0, 54)[1]
plt.plot(x_est_MAE, y_est_MAE, color = 'red', linewidth = 3)
```

На основе полученных графиков можно сделать вывод, что оценка методом MAE является более устойчивой к выбросам, чем оценка методом MSE.

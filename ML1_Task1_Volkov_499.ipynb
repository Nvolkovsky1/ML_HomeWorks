```python
from sklearn import model_selection, datasets, metrics, neighbors 
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split

%pylab inline
```

```python
#cгенерируем обучающую выборку и визуализируем ее в координатах признаков
sample = datasets.make_classification(n_samples=1000, n_features = 2, n_informative = 2, 
                                                      n_classes = 4, n_redundant=0, 
                                                      n_clusters_per_class=1, random_state=5)
```

```python
print sample[0][:10]
print sample[1][:10]
```

```python
colors = ListedColormap(['red', 'blue', 'green','yellow'])
light_colors = ListedColormap(['lightcoral', 'lightblue', 'lightgreen', 'lightyellow'])
plt.figure(figsize=(8,6))
#plt.scatter(map(lambda x:x[0], sample[0]), map(lambda x: x[1], sample[0]))
plt.scatter(map(lambda x: x[0], sample[0]), map(lambda x: x[1], sample[0]), 
              c=sample[1], cmap=colors, s=100)
```

```python
#разделим выборку на тренировочную и тестовую части 
X_train, X_test, Y_train, Y_test = train_test_split(sample[0], sample[1], test_size=0.4, random_state=9)
print X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
def get_meshgrid(data, step=.05, border=.5,):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
```

```python
def plot_decision_surface(n_neighbors, X_train, Y_train, X_test, Y_test, 
                          colors = colors, light_colors = light_colors):

    estimator = neighbors.KNeighborsClassifier(n_neighbors)

    #fit model
    estimator.fit(X_train, Y_train)
    
    #set figure size
    pyplot.figure(figsize = (16, 6))
    
    #plot decision surface on the train data 
    pyplot.subplot(1,2,1)
    xx, yy = get_meshgrid(X_train)
    mesh_predictions = np.array(estimator.predict(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
    pyplot.pcolormesh(xx, yy, mesh_predictions, cmap = light_colors)
    pyplot.scatter(X_train[:, 0], X_train[:, 1], c = Y_train, s = 100, cmap = colors)
    pyplot.title('Train data, k=%d, accuracy=%.2f' % (n_neighbors, metrics.accuracy_score(Y_train, estimator.predict(X_train))))
    
    #plot decision surface on the test data
    pyplot.subplot(1,2,2)
    pyplot.pcolormesh(xx, yy, mesh_predictions, cmap = light_colors)
    pyplot.scatter(X_test[:, 0], X_test[:, 1], c = Y_test, s = 100, cmap = colors)
    pyplot.title('Test data, k=%d, accuracy=%.2f' % (n_neighbors, metrics.accuracy_score(Y_test, estimator.predict(X_test))))
```

```python
#Проведем кластеризацию методом kNN при различных k, визуализируем разделющие кривые и посчитаем 
#accuracy для каждого значения k.

for k in xrange(1, 51, 5):
    plot_decision_surface(k, X_train, Y_train, X_test, Y_test)
```

```python
kf = KFold(n_splits = 5)
    
def accuracy(n_neighbors):
    mean = 0
    for train_index, test_index in kf.split(sample[0], sample[1]):
       X_train, X_test = sample[0][train_index], sample[0][test_index]
       Y_train, Y_test = sample[1][train_index], sample[1][test_index]
       estimator = neighbors.KNeighborsClassifier(n_neighbors)
       estimator.fit(X_train, Y_train)
       mean += estimator.score(X_test, Y_test)
    return float(mean) / 5
```

```python
#Теперь, исползуя метод kFolf для более точной оценки качества кластеризации, построим график зависимости
#accuracy от количества рассматриваемых соседей.
results = [accuracy(k) for k in xrange(1, 101)]

print 'Оптимальное количество рассмтариваемых соседей: %d \n При этом достигается accuracy: %.3f' % (\
results.index(max(results)), max(results))
plt.figure(figsize = (10, 6))

plt.scatter(xrange(1, 101), results, s = 50)
plt.title('Accuracy(#neighbors)')
plt.xlabel('#neighbors')
plt.ylabel('Accuracy')
    
```

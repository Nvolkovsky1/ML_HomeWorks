```python
#загрузим датасет digits и выведем первую строчки признаков
digits = datasets.load_digits()
digits_frame = pd.DataFrame(digits.data)
digits_frame.head()
```

```python
X_digits = digits.data
Y_digits = digits.target
```

```python
def get_mean(estimator, X, Y):
    return sklearn.model_selection.cross_val_score(estimator, X, Y).mean()
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
estimators = [BernoulliNB(), MultinomialNB(), GaussianNB()]

print 'BernoulliNB :', cross_val_score(estimators[0], digits.data, digits.target).mean()
print 'MultinomialNB :', get_mean(estimators[1], X_digits, Y_digits)
print 'Gaussian :', get_mean(estimators[2], X_digits, Y_digits)
```

Видим, что для датасета digits лучший результат дает использование MultinomialNB.

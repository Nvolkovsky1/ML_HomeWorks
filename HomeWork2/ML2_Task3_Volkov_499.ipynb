### Вспомогоательные функции

```python
#какой ответ даём в листе
def answer(list_vertexes):
    return np.mean(list_vertexes)

#MSE
def mse(arr_1, arr_2):
    arr_1 = np.array(arr_1)
    arr_2 = np.array(arr_2)
    return np.sum((arr_1 - arr_2)**2)

#функция ошибки, которой будем оценивать качество разбиения
def estimate_err_1(left, right):
    H_L = mse(left[:, left.shape[1] - 1], answer(left[:, left.shape[1] - 1]))
    H_R = mse(right[:, right.shape[1] - 1], answer(right[:, right.shape[1] - 1]))
    l_len = left.shape[0]
    r_len = right.shape[0]
    sum_len = l_len + r_len                   
    return float(l_len * H_L + r_len * H_R) / sum_len

def estimate_err_2(left, right):
    l_len = left.shape[0]
    r_len = right.shape[0]                   
    return (r_len - l_len)**2 + estimate_err_1(left, right)

#сетка,по которой выбираем порог
def threshold_greed_1(X, feature, frequency = 1):
    min_val = min(X[:,feature])
    max_val = max(X[:,feature])
    step = float(max_val - min_val) / (X.shape[0] * frequency)
    greed = np.array([min_val + k * step for k in xrange(frequency * X.shape[0])])
    return greed

def threshold_greed_2(X, feature):
    min_val = min(X[:,feature])
    max_val = max(X[:,feature])
    dif = max_val - min_val
    greed = np.array([min_val + dif * random() for k in xrange(X.shape[0])])
    return greed

def threshold_greed_3(X, feature):
    greed = X[:, feature] 
    return greed
```

```python
class Node:
    def __init__(self, max_depth = 128):
        self.max_depth = max_depth
        self.left = None
        self.right = None
        self.feature = None
        self.threshold = None
        self.answer = None
    
    def fit(self, X, y, max_depth):
        X = np.array(X)
        y = np.array(y)
        if max_depth == 0:
            self.answer = answer(y)
            return 
        
        min_error = None
        left = None
        right = None
        
        data = np.concatenate([X, y.reshape((-1, 1))], axis=1)
        for feature in xrange(X.shape[1]):
            greed = threshold_greed_3(X, feature)
            for threshold in greed:
                    left = data[data[:, feature] <= threshold]
                    right = data[data[:, feature] > threshold]
                    error = estimate_err_1(left, right)
                    if  error < min_error or min_error == None: 
                        min_error = error
                        self.feature = feature
                        self.threshold = threshold
        if left.shape[0] > 0 and right.shape[0] > 0:
            self.left = Node()
            self.left.fit(left[:, 0 : left.shape[1] - 1], left[:, left.shape[1] - 1], max_depth - 1)
            self.right = Node()
            self.right.fit(right[:, 0 : right.shape[1] - 1], right[:, right.shape[1] - 1], max_depth - 1)
        else:
            self.answer = answer(y)
            return 
    
    def predict(self, list_of_features):
        if self.answer != None:
            return self.answer
        if list_of_features[self.feature] <= self.threshold:
            return self.left.predict(list_of_features)
        else:
           return self.right.predict(list_of_features)
        
```

```python
class DecisionTree:
    def __init__(self, max_depth = 128):
        self.max_depth = max_depth
        root = None
    def fit(self, X, y):
        self.root = Node()
        self.root.fit(X,y, self.max_depth)
    def predict(self, X):
        target = [self.root.predict(features) for features in X]
        return target
```

```python
boston = sklearn.datasets.load_boston()
data = boston.data
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.25, random_state = 57)
print X_train.shape, y_train.shape
```

Используем равномерный проход по возможным значениям порога, т.е. используем функцию threshold_greed_1. Для оценки качества разбиения в вершинке используем функцию err_estimator_1.

```python
DTree = DecisionTree(3)
DTree.fit(X_train, y_train)
predictions = DTree.predict(X_test)
print mse(predictions, y_test) / float(len(predictions))
```

```python
DTree = DecisionTree(10)
DTree.fit(X_train, y_train)
predictions = DTree.predict(X_test)
print mse(predictions, y_test) / float(len(predictions))
```

```python
train_err = []
test_err = []
limit = 50
for depth in xrange(1, limit):
    clf = DecisionTree(depth)
    clf.fit(X_train, y_train)
    
    predictions_tn = clf.predict(X_train)
    train_err.append(mse(predictions_tn, y_train) / (100 * float(len(predictions_tn))))
    
    predictions_tt = clf.predict(X_test)
    test_err.append(mse(predictions_tt, y_test) / (100 * float(len(predictions_tt))))
        
plt.figure(figsize = (18, 6))
    
plt.subplot(1,2,1)
plt.plot(range(1,limit), train_err, color = 'lightblue')
plt.scatter(range(1,limit), train_err, color = 'blue')
plt.title('Train data')
plt.xlabel('max depth')
plt.ylabel('cross val score')

plt.subplot(1,2,2)
plt.plot(range(1,limit), test_err, color = 'lightblue')
plt.scatter(range(1,limit), test_err)
plt.title('Test data')
plt.xlabel('max depth')
plt.ylabel('cross val score')
```

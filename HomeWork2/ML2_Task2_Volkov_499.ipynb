```python
from sklearn import tree, model_selection as ms
from sklearn.model_selection import  cross_val_score
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#скачаем нужные файлы
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.doc
```

### Представим данные в удобном виде.

```python
import re
names = []
convert = {}
with open("german.doc") as f:
    while True:
        l = f.readline()
        if not l:
            break
        l = l.strip()
        if re.match("Attr?ibute \d+", l):
            names.append(f.readline().strip())
        match = re.match("(A\d+) : (.*)", l)
        if match:
            convert[match.group(1)] = match.group(2)
            
```

```python
data = pd.read_csv("german.data", ' ', skipinitialspace=True, names=names+["target"]).replace(convert)
```

```python
data[:5]
```

#### В получившейся табличке не все признаки числовые. Исправим это с помощью функции pd.get_dummies.

```python
pd.options.display.max_columns = 200
credits = pd.get_dummies(data)
credits.head()
```

#### Теперь построим решающее дерево и визуализируем его.

```python
X = credits.drop("target", 1)
Y = credits["target"]
X_train, X_test, Y_train, Y_test = ms.train_test_split(X, Y, test_size=0.4, random_state=9)
```

```python
clf = tree.DecisionTreeClassifier("entropy", max_depth=3, random_state=57)
clf.fit(X_train, Y_train)
    
dot_data = tree.export_graphviz(clf, out_file="small_tree.out", 
                         feature_names=credits.columns,  
                         class_names=["good", "bad"],  
                         filled=True, rounded=True,  
                         special_characters=False)
```

```python
import pydotplus 
from IPython.display import Image

graph = pydotplus.graphviz.graph_from_dot_file("small_tree.out")  
Image(graph.create_png())
```

## Оценим качество на обучающей и тестировочной выборке.

```python
print "Обучающая выборка %f" % np.mean(cross_val_score(clf, X_train, Y_train))
print "Тестировочная     %f" % np.mean(cross_val_score(clf, X_test, Y_test))
```

### Посмотрим на зависимость качества классифакации от глубины дерева.

```python
train_err = []
test_err = []
limit = 35
for depth in xrange(1, limit):
    clf = tree.DecisionTreeClassifier("entropy", max_depth=depth, random_state=59)
    clf.fit(X_train, Y_train)
    train_err.append(np.mean(cross_val_score(clf, X_train, Y_train)))
    test_err.append(np.mean(cross_val_score(clf, X_test, Y_test)))

plt.figure(figsize = (18, 6))
    
plt.subplot(1,2,1)
plt.plot(range(1,limit), train_err, color = 'lightblue')
plt.scatter(range(1,limit), train_err, color = 'blue')
plt.title('Train data')
plt.xlabel('max depth')
plt.ylabel('cross val score')

plt.subplot(1,2,2)
plt.plot(range(1,limit), test_err, color = 'lightblue')
plt.scatter(range(1,limit), test_err)
plt.title('Test data')
plt.xlabel('max depth')
plt.ylabel('cross val score')
    
```

Видим, что с увеличением слоев у дерева, качество предсказания уменьшается. Скорее всего начинается переобучение.
